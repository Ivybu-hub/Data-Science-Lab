```python
#import Matplotlib, pandas, and plotly
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#Task 1.5.1
df1 = pd.read_csv('brasil-real-estate-1.csv', encoding='latin1')
```

```python
#Inspect dataFrame df1 using the 'info' and 'head' method
df1.info()
```

```python
# Task 1.5.2 Drop all rows with NaN values from the DataFrame df1
# Checking for null/missing values
df1.isnull().sum()
```

```python
df1 = df1.dropna()
```

```python
# Task 1.5.3 Use the "lat" and "lon" columns directly, as "lat-lon" does not exist in df1.
# Ensure that the data type is float
df1['lat', 'lon'] = pd.to_numeric(df1['lat-lon'], errors='coerce')
```

```python
df1['lat'] = pd.to_numeric(df1['lat'])
df1['lon'] = pd.to_numeric(df1['lon'])
```

```python
#Task 1.5.4 Use the "place_with_parent_names" column to create a "state"
df1["state"]= df1["place_with_parent_names"].str.split("|", expand=True) [2]
```

```python
df1['lat'] = pd.to_numeric(df1['lat'])
```

```python
# Task 1.5.5 Transform the "price_usd" column of df1 so that all values are floating-point numbers instead of strings
# Strip leading/trailing spaces from column names
df1.columns = df1.columns.str.strip()

# Transform the "price_usd" column of df1 so that all values are floating-point numbers instead of strings
df1["price_usd"] = df1["price_usd"].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
```

```python
# Task 1.5.6 Drop "lan-lon" and "place_with_parent_names" columns from df1
df1 = df1.drop(['place_with_parent_names'], axis = 1)
df1 = df1.drop(['lat-lon'], axis=1)
```

```python
# Task 1.5.7 Import the CSV file brasil-real-estate.csv into the DataFrame df2
df2 = pd.read_csv("brasil-real-estate-1.csv", encoding='latin1')
```

```python
df2.info()
```

```python
df2.head()
```

```python
df2['state'].unique()
```

```python
# Task 1.5.8 Use the 'price_brl' column to create a new column named 'price_usd'. Keep in mind that when this data was collected in 2015 and 2016, a US dollar cost 3.19 Brazilian reals.
df2['price_usd'] = df2['price_brl']/3.19
```

```python
df2['price_usd'] = pd.to_numeric(df2['price_usd'])
```

```python
#Task 1.5.9 Drop the 'price_brl' column df2, as well as any rows that have NaN values
df2 = df2.drop(['price_brl'], axis = 1)
```

```python
df2 = df2.dropna()
```

```python
#Task 1.5.10 Concantenate df1 and df2 to create a new DataFrame named df.
# Task 1.5.10 Concatenate df1 and df2 to create a new DataFrame named df.
frames = [df1, df2]
df = pd.concat(frames)
print("df shape:", df.shape)
```

```python
df.head()
```

```python
df['region'].unique()
```

```python
#Explore
#Complete the code to create a scatter_mapbox showing the location of the properties in df
fig =px.scatter_mapbox(
    df,
    lat= df['lat'],
    lon= df['lon'],
    center={'lat': -14.2, 'lon': -51.9},
    width=600,
    height=600,
    hover_data=['price_usd'], # Display price when hovering mouse over house
)

fig.update_layout(mapbox_style="open-street-map")

fig.show()
```

```python
#Build histogram
plt.hist(df['price_usd'])

#Label axes
plt.xlabel ("Price[USD]")
plt.ylabel ("Frequency")

#Add title
plt.title("Distribution of House Prices in Brazil")
```

```python

#Build boxplot
plt.boxplot(df['area_m2'])

#Label x-axis
plt.xlabel ("Area [square meters]")

#Add title
plt.title("Distribution of House Areas in Brazil")
```

```python
mean_price_by_region = df.groupby('region')['price_usd'].mean().sort_values(ascending=True)
mean_price_by_region
```

```python
#Build bar chart, label axes, and add title
mean_price_by_region.plot(
    kind='bar',
    xlabel='Region',
    ylabel='Mean Price [USD]',
    title='Mean Price by Region in Brazil')
```
